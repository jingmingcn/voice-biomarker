```python
# Plot the time series
plt.rcParams['font.sans-serif']=['Songti SC']  #解决中文显示乱码问题
plt.rcParams['axes.unicode_minus']=False  #解决坐标轴负数的负号显示问题
plt.style.use('fivethirtyeight')
```

```python
folder = 'Feature-7-Image'

for name in names:
    ax = df.boxplot(column=[name+'_mean',name+'_std',name+'_min',name+'_25',name+'_50',name+'_75',name+'_max'],by='y',grid=True,fontsize=10,layout=(1,7),figsize=(6,6),return_type='axes')
    ax[name+'_mean'].get_figure().suptitle(name)
    ax.apply(lambda x:x.set_title(x.get_title().split('_')[-1]))
    ax.apply(lambda x:x.set_xlabel(''))
    plt.savefig(folder+'/'+name+'.png')
```
